## Objective
The objective of this task is to perform exploratory data analysis on the Iris dataset, understand feature distributions, identify patterns, and visualize relationships between variables.

```python
# Import necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris

# Set style for better visuals
sns.set(style="whitegrid")
```

```python
# Load Iris dataset
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Map numeric species to actual names
df['species'] = df['species'].map({
    0: 'Setosa',
    1: 'Versicolor',
    2: 'Virginica'
})

# Display first 5 rows
df.head()
```

## Dataset Overview

In this section, we examine the structure of the dataset, including shape, data types, and missing values.

```python
# Shape of dataset
print("Shape of dataset:", df.shape)

# Data types
df.info()

# Check for missing values
df.isnull().sum()
```

## Summary Statistics

We analyze the central tendency and dispersion of numerical features.

```python
df.describe()
```

## Feature Distributions

We visualize the distribution of numerical features to understand their spread and patterns.

```python
# Histograms of numerical features
df.hist(figsize=(10,8))
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

## Boxplots by Species

Boxplots help identify differences in feature distributions across species.

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="species", y="petal length (cm)")
plt.title("Petal Length by Species")
plt.show()
```

## Correlation Analysis

We analyze relationships between numerical features using a correlation matrix.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop("species", axis=1).corr(), 
            annot=True, 
            cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Pairwise Relationships

Pairplot visualizes relationships between all numerical features across species.

```python
sns.pairplot(df, hue="species")
plt.show()
```

## Key Observations

1. Petal length and petal width show strong positive correlation.
2. Setosa species is clearly separable from the other two species.
3. Versicolor and Virginica show some overlap but can still be distinguished using petal features.
4. Sepal width has weaker correlation with other features.
5. No missing values are present in the dataset.

## Conclusion

The exploratory data analysis reveals that petal-related features are the most significant in distinguishing iris species. 
The dataset is clean, well-structured, and suitable for building classification models.

Overall, the Iris dataset demonstrates clear separability between classes, particularly for the Setosa species.
